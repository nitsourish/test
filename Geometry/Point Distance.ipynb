```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1,y1 = 0,0
x2,y2 = 250,250
n = 25
Xs = np.arange(0,250,n)
Ys = np.random.randint(0,250, len(Xs))

print('X:%s\nY:%s' %(Xs,Ys))
```

```python
# When Y=X on diagonal line
Ys-Xs
# equals distance to line, and indicates +-
```

```python
def get_distance(lx1,lx2,ly1,ly2,Px,Py):
    #linex,linex,liney,liney,pointx,pointy
    distance = abs((ly2-ly1)*Px-(lx2-lx1)*Py+lx2*ly1-ly2*lx1) / np.sqrt((ly2-ly1)**2 + (lx2-lx2)**2)
    return distance
```

```python
Ds = []
for Px,Py in zip(Xs,Ys):
    Ds.append(get_distance(x1,x2,y1,y2,Px,Py))
Ds
```

```python
plt.figure(figsize=(10,10))

plt.plot([x1,x2], [y1,y2], 'k-')
er=30
plt.plot([x1,x2], [y1-er,y2-er])
plt.plot([x1,x2], [y1+er,y2+er])

labels = np.arange(0,len(Xs))
plt.scatter(Xs, Ys, label=labels, s=Ds*100)

plt.xlim(0,250)
plt.ylim(0,250)
```
